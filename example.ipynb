## تنظیمات

### تنظیمات اندازه

یکی از اولین چیزهایی که می‌توان تغییر داد اندازه صفحه تقویم است.
این اندازه وارد شده اندازه اصلی فایل است  که به اصطاح به آن اندازه بعد از برش می‌گویند.

این اندازه دو پارامتر `width` و `height` تعریف می‌شود که باید بر اساس میلیمتر آن را وارد کرد.

چند تا از اندازه‌های متعارف در زیر آورده شده:

</div>


````python
standardSizes = {
    "A4": {'width': 210, 'height': 297},
    "A5": {'width': 148, 'height': 210},
    "A6": {'width': 105, 'height': 148},

    "B5": {'width': 176, 'height': 250},
    "B6": {'width': 125, 'height': 176},

    "17x23": {'width': 170, 'height': 230},  # قطع وزیری
    "14x21": {'width': 140, 'height': 210},  # (مناسب چاپ دو صفحه کنار هم روی A4) قطع رقعی

    "11.5x16.5": {'width': 115, 'height': 165},  # قطع جیبی بزرگ
    "10x14": {'width': 100, 'height': 140},  # (مناسب چاپ دو صفحه کنار هم روی A5) قطع جیبی کوچک

    "11.5x22": {'width': 115, 'height': 220},  # قطع پالتویی بزرگ
    "11x19": {'width': 110, 'height': 190},  # قطع پالتویی کوچک
}
````


<div dir="rtl"font-family="vazirmatn" >


**توجه:** 

بهتر است برای اینکه بتوان روی کاغذ A4 یا A5 چاپ کرد از سایز کوچکتری از کاغذ استفاده کرد و حاشیه رنگ را برید تا به تمیزی بیشتری در لبه‌ها رسید و لبه سفید حاشیه کاغذ که پرینتر امکان چاپ روی آن ندارد را در فایل نداشت. 

برای این تقویم با توجه به اینکه خود طراحی حاشیه خالی در بالا و پایین دارد نیازی به تغییر `height` نیست و فقط کافی است `width` از هر طرف به میزان تقریبی ۵ میلیمتر کوچکتر در نظر گرفته شود.

در ادامه جزئیات مربوط به چاپ کامل‌تر توضیح داده می‌شود.

</div>


````python
pageSize = "14x21"

setting.update(standardSizes[pageSize])
````

### تنظیمات خطوط

با توجه به اینکه ساختار اصلی تقویم بر اساس دفتر خط‌دار طراحی شده یکی دیگر از تنظیمات مهم می‌تواند این مورد باشد.

فاصله بین خطوط افقی با پارامتر `lineHeight` تعریف می‌شود که براساس میلیمتر است.

از آنجایی که ساختار اصلی طراحی تقویم هفتگی است یکی دیگر از پارامتر‌ها `daysHeight` است که نماینده تعداد خط‌های یک روز است. پیشنهاد می‌شود برای برهم نخوردن ساختار کلی طراحی و نمایش این پارامتر را بیش از ۳ قرار دهید.

برای مثال در تصویر زیر از پارامتر‌های هرکدام به ترتیب:

</div>



![](src/img/lineHeight.svg)


```json
"left":   {"lineHeight": 5 ,"daysHeight": 5},
"middle": {"lineHeight": 6 ,"daysHeight": 4},
"right":  {"lineHeight": 8 ,"daysHeight": 3},
```


<div dir="rtl"font-family="vazirmatn" >

توجه داشته باشید که تنظیم این پارامتر با ارتفاع صفحه ارتباط مستقیمی دارد. لازم است که فضای مورد نیاز برای نمایش همه صفحه در صفحه هفتگی وجود داشته باشد و در نتیجه لازم است.
همچنین در رابطه با حاشیه سفید در ادامه توضیح خواهد داده شد.

</div>


`lineHeight` * `daysHeight` * 7 + حاشیه سفید مورد نیاز ≤ `height`


````python
setting["lineHeight"] = 5
setting["daysHeight"] = 5
````


<div dir="rtl"font-family="vazirmatn" >

(عکس سمت چپ)

با این پارامتر می‌توان تقویمی طراحی کرد که نیمه بالایی صفحه خط‌دار ساده و نیمه پاییتی تقویم باشد

</div>


![](src/img/lineShiftDown.svg)

### تنظیمات حاشیه سفید

برای تنظیم حاشیه سفید از پارامتر `padding` استفاده میشود که به عنوان ورودی یک لیست چهارتایی از اعداد می‌گیرد.

با توجه به اینکه در طراحی دو صفحه‌ای padding راست و چپ با توجه به تفاوت داشتن صفحه راست و چپ معنی دقیق ندارد. هرکدام از عددها با ترتیب حاشیه بالایی، حاشیه میانی، حاشیه پایینی و حاشیه بیرونی را نمایندگی می‌کند.
</div>

`[#top, #outside, #bottom, #inside]`

<div dir="rtl"font-family="vazirmatn" >

همچنین padding را می‌توان به صورت دیکشنری نیز تعریف کرد.

</div>

`{'top': 5, 'outside': 5, 'bottom': 5, 'inside': 0}`

<div dir="rtl"font-family="vazirmatn" >

#### نحوه مناسب محاسبه اعداد برای حاشیه

در ابتدا پیشنهاد می‌شود برای اینکه طرح کلی بهم نریزد حداقل حاشیه بالا و پایین را برابر با `lineHeight` قرار دهید.

حال میزان باقی‌مانده از فضای عمودی را می‌توانید به دلخواه بین حاشیه بالا و پایین تقسیم کرده و به این اعداد اضافه کنید.

میزان فضای باقی‌مانده برابر با:

</div>

remianSpace = `height` - `lineHeight` \* `daysHeight` \* 7 + 2 \* `lineHeight` 

<div dir="rtl"font-family="vazirmatn" >

**توجه:** در صورت استفاده از `lineShiftDown` آن را در محاسبات خود در نظر بگیرید.


بعد از تقسیم این میزان باقی‌مانده پیشنهاد می‌شود از حاشیه پایین یک میلیمتر کم کنید تا در رندر شدن و هماهنگ شدن دیگر صفحات  با صفحه هفتگی به مشکل نخورید.

برخلاف حاشیه بالا و پایین حاشیه درونی و بیرونی اختیاریست و بیشتر جنبه طراحی دارد.  و باعث فاصله گرفتن طراحی به خصوص خطوط افقی از سمت راست و چپ کاغذ می‌شود. (عکس سمت راست) این حاشیه‌ها با صورت پیش‌فرص `0` هستند. (عکس سمت چپ)

</div>


![](src/img/padding.svg)


````python
padding = {'top': 25, 'bottom': 9, 'inside': 0, 'outside': 0}

setting["padding"] = padding
````
